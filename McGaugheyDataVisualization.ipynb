### Transformations

In order to work with the data more effectively, I wanted to convert the participants’ average time on social media from values of type string to values of type float. I used the numeric average of each value to do so. For example, I changed “Between 2 and 3 hours” to 2.5. Having numeric values for this data allowed me to visualize correlations in a more effective way. Additionally, for consistency, I used find and replace within the actual file to consolidate gender identities. For example, I renamed “NB” to “Non-binary”. 

### Visualizations

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("smmh.csv")
```

```python
pd.set_option('future.no_silent_downcasting', True)
df.replace(to_replace = "Less than an Hour", value = 0.5, inplace = True)
df.replace(to_replace = "Between 1 and 2 hours", value = 1.5, inplace = True)
df.replace(to_replace = "Between 2 and 3 hours", value = 2.5, inplace = True)
df.replace(to_replace = "Between 3 and 4 hours", value = 3.5, inplace = True)
df.replace(to_replace = "Between 4 and 5 hours", value = 4.5, inplace = True)
df.replace(to_replace = "More than 5 hours", value = 5.5, inplace = True)
```

#### The following visualizations depict a strong correlation between different social media behaviors and different mental health tendencies. 

```python
sns.kdeplot(data=df, x = "average_time_social_media", y = "depressed_or_down", fill = True, cmap = "mako_r")
```

_Individuals spending more time on social media tend to feel more depressed. Further, we can imagine that those who feel depressed try to fill their time with numbing distractions such as scrolling. The cycle continuously feeds itself._

```python
sns.kdeplot(data=df, x = "bothered_by_worries", y = "comparisons_to_others_through_social_media", fill = True, cmap = "mako_r")
```

_Individuals who compare themselves to others on social media tend to be most bothered by worries. The anxiety related to not feeling worthy is compounded by the presence of so many individuals online who do seem worthy._

#### The following visualization represents how much participatns compare themselves to others on social media by gender.

```python
count = sns.countplot(data=df, x = "comparisons_to_others_through_social_media", hue = "gender", palette = "Paired")
sns.move_legend(count, "upper left", bbox_to_anchor=(1, 1))
```

_Though levels for males and females remain consistent in the lower categories, a significantly larger number of females rank their levels of comparisons through the use of social media as high as possible. The expectations placed upon women in our society, particularly in the realm of appearance, loom largely as we try to be our most authentic selves. On social media, there is an endless supply of examples of how we should be, and there is always someone who seems to be doing better._

#### The following visualization depicts time spent on socials based on relationship status.

```python
sns.set_palette("pastel")
```

```python
hist = sns.histplot(data = df, x = "average_time_social_media", hue = "relationship_status")
sns.move_legend(hist, "upper left", bbox_to_anchor=(1, 1))
```

_Of those who spend the most time on social media, many are single. Social media can feel like an engaging distraction or even a presence that helps to fill a void. Those without a significant other might spend more time trying to fill the empty space, or even comparing themselves to others to openly broadcast their happy relationships._

#### The following visualizations relate to levels of concentration/focus and how they may be impacted by social media usage. 

```python
sns.lineplot(data = df, x = "social_media_without_specific_purpose", y = "distracted_by_social_media_when_busy")
```

_There is a strong correlation between social media usage for no particular reason and social media being a distraction. When using social media for no specific purpose, one can get lost in the dissociative activity of mindlessly scrolling. This activity can become a habit that beckons the user even when they are busy with something else._

```python
facet = sns.FacetGrid(data = df, col = "occupation")
facet.map(sns.countplot, "difficult_to_concentrate", order = "12345")
```

_This visualization depicts difficulty concentrating by occupation. Evidently, most university students who participated rank their difficulty concentrating very highly. Many factors affect young college students today. We are busier than we have been in our lives and have so much to concentrate on. With so many racing thoughts, it can be difficult to hone in on just one thing, and technology certainly plays a part in the restlessness of our minds._

### Visualization Approach

In order to make my visualizations, I mainly wanted to focus on what was strongly correlated. I spent some time generating every visualization I could possibly think of, then went through and chose which ones I felt most strongly demonstrated my arguments. I chose KDE plots for values which had such a strong correlation that the visual aspect completely encapsulated the story. Where count was more important than correlation, I went with count plots and used different colors to represent different demographics. I used one line plot to show a more simple and straightforward correlation. And finally, I used a facet map when showing occupation in relation to difficulty focusing. I found this technique particularly useful because it allows viewers to see a few charts, side by side, and make a generalization about all the factors involved.
